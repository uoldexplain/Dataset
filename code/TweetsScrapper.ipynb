```python
def ScrapTweets(start,end,t):
    attributes_container = []
    criteria = f" since:{start} until:{end} lang:ur"
    print(criteria)
    count=0
    try:
        for i,tweet in enumerate(sntwitter.TwitterSearchScraper(criteria,
            'mode = snscrape.modules.twitter.TwitterSearchScraperMode.TOP').get_items()):
            attributes_container.append([tweet.id,tweet.content, tweet.date, tweet.user.username])
            time.sleep(0.1)
            count=i
            if count>10: break
            if count>0 and count%100==0:
                clear_output()
                print("Scrap Count: ",i," At ", datetime.now().time(), "------Day-----",start ," started At ",t)
        #end of for
        if count>0:            
            tweets_df = pd.DataFrame(attributes_container, columns=["ID", "Tweet","Date","User"])
            tweets_df['Date'] = tweets_df['Date'].dt.tz_localize(None)
            tweets_df.to_excel("tweets-{}-{}.xlsx".format(start,count), sheet_name = "{}".format(start))
    
            print("done",count)
        print("Code executed successfully with count= {}".format(count))
    except Exception as e:
        print("Except Code started successfully",e)
        if count>0:
            tweets_df = pd.DataFrame(attributes_container, columns=["ID", "Tweet","Date","User"])
            tweets_df['Date'] = pd.to_datetime(tweets_df['Date'], errors='coerce')
            tweets_df['Date'] = tweets_df['Date'].dt.tz_localize(None)
            tweets_df.to_excel("tweets-{}-{}.xlsx".format(start,count), sheet_name = "{}".format(start))
            print("done",count)
        print("Except Code executed successfully")
    finally:
        print("Finally Executed")
#end of function____________________________
```

```python
#Search day based Urdu Tweets with start and End dates
from IPython.display import clear_output
from datetime import datetime
import pandas as pd
#11:32
list=["2022-12-24","2022-12-25"]#,"2022-12-26","2022-12-27","2022-12-28","2022-12-29","2022-12-30","2022-12-31"]
for i in range(len(list)-1):
    start=list[i]
    end=list[i+1]
    print(start,"____started_______")
    ScrapTweets(start,end,datetime.now().time())
```
